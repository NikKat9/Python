```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.ones((5,),dtype='float128')
x.dtype
```

```python
lenght=0.2
width=0.3
X, Y = np.meshgrid(np.linspace(0,lenght,500,dtype=np.longdouble),np.linspace(width,0,500,dtype=np.longdouble))
```

```python

T_analytical = sum(((np.sinh(n*np.pi*Y/lenght,dtype='float128')/np.sinh(np.pi*n*width/lenght,dtype='float128'))*(2/(n*np.pi))*(pow(-1,n+1)+1)*np.sin(n*np.pi*X/lenght)) for n in range (1,1000))
ak=1
```

```python
T_analytical.dtype
```

```python
plt.figure(dpi=720)
plt.setp(plt.axes().spines.values(), linewidth=3)
plt.xlabel(' X (m) ',fontsize=12,fontweight='bold')
plt.ylabel('Y (m) ',fontsize=12,fontweight='bold')
plt.xticks(np.linspace(0,lenght,5))
if ak==1:
    T_analytical=T_analytical*(100-20) + 20 
    ak=ak+1
plt.contourf(X,Y,T_analytical, alpha=1, cmap='rainbow',vmax=100.0000)
plt.clim(20,100)
plt.colorbar()
plt.savefig('Analytic')
```
